```python
from imdb import Cinemagoer

# create an instance of the Cinemagoer class
ia = Cinemagoer()

# get a movie
movie = ia.get_movie('0133093')

# print the names of the directors of the movie
print('Directors:')
for director in movie['directors']:
    print(director['name'])

# print the genres of the movie
print('Genres:')
for genre in movie['genres']:
    print(genre)

# search for a person name
people = ia.search_person('Mel Gibson')
for person in people:
   print(person.personID, person['name'])
```

```python
from imdb import Cinemagoer
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# Set plot style
rcParams.update({'font.family': 'serif',
                  'font.serif': ['Times'],
                  'font.size': 12,
                  'axes.labelsize': 12,
                  'axes.titlesize': 16,
                  'xtick.labelsize': 12,
                  'ytick.labelsize': 12,
                  'legend.fontsize': 12})

# Create an instance of the IMDb class
ia = IMDb()

# Get the top-rated Indian movies
movies = ia.get_top250_indian_movies()

# Get the movies released in 2020 and 2021
movies_2020_2021 = [movie for movie in movies if 'year' in movie and movie['year'] in [2020, 2021]]

# Get the movies released in all other years
movies_other_years = [movie for movie in movies if 'year' in movie and movie['year'] not in [2020, 2021]]

# Calculate the average rating for the two sets of movies
avg_rating_2020_2021 = np.mean([movie['rating'] for movie in movies_2020_2021])
avg_rating_other_years = np.mean([movie['rating'] for movie in movies_other_years])

# Calculate additional movie statistics
total_movies_2020_2021 = len(movies_2020_2021)
min_rating_2020_2021 = np.min([movie['rating'] for movie in movies_2020_2021])
max_rating_2020_2021 = np.max([movie['rating'] for movie in movies_2020_2021])
std_rating_2020_2021 = np.std([movie['rating'] for movie in movies_2020_2021])

total_movies_other_years = len(movies_other_years)
min_rating_other_years = np.min([movie['rating'] for movie in movies_other_years])
max_rating_other_years = np.max([movie['rating'] for movie in movies_other_years])
std_rating_other_years = np.std([movie['rating'] for movie in movies_other_years])

# Plot the results
labels = ['2020-2021\n' + f'({total_movies_2020_2021} movies)', 'Other years\n' + f'({total_movies_other_years} movies)']
values = [avg_rating_2020_2021, avg_rating_other_years]

fig, ax = plt.subplots(figsize=(8,6))
ax.bar(labels, values, color=['#0077B6', '#FFC300'])
ax.set_title('Average Rating Of Top 250 Indian movies IMBD Scores', fontweight='bold')
ax.set_ylabel('Average rating', fontweight='bold')
ax.set_ylim(7.5, 8.2)

# Add text to show additional movie statistics
ax.text(0, avg_rating_2020_2021+0.005, f'Min rating: {min_rating_2020_2021:.1f}\nMax rating: {max_rating_2020_2021:.1f}\nStd rating: {std_rating_2020_2021:.2f}', ha='center', fontweight='bold')
ax.text(1, avg_rating_other_years+0.005, f'Min rating: {min_rating_other_years:.1f}\nMax rating: {max_rating_other_years:.1f}\nStd rating: {std_rating_other_years:.2f}', ha='center', fontweight='bold')

# Add grid lines and background color
ax.grid(True, axis='y', linestyle='--', alpha=0.5)
fig.patch.set_facecolor('#F2F2F2')

plt.show()

```
